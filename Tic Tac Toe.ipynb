```python
#                                 TIC-TAC-TOE 

# Global Variables
# It is always a good practice to assign the global variables using a return value from a function
# If we change a global variable in the function the visibility of the change is obscured 

list_position = ['','','','','','','','','']
player = 'Z'
winner = ''
tac_count = 0

# ----------------------------------------------------------------------------------------------------------------
def print_board (position=0):
    '''
    Print the board with the position the player chooses
    '''
    
    
    if position >0 and position<10 : # check for position in range 0-9
        
        list_position [position-1] = player
        #tac_count=+1 - Local variable 
        print (f'\n print_board: [ player: {player} \t tac_count: {tac_count} \t list_position: {list_position}]')
    
    print ('\n\n')
    print (f"   {list_position[6]}   |  {list_position[7]}   | {list_position[8]} ")    
    print (f"------------------")
    print (f"   {list_position[3]}   |  {list_position[4]}   | {list_position[5]} ")    
    print (f"------------------")
    print (f"   {list_position[0]}   |  {list_position[1]}   | {list_position[2]} ") 
    print ('\n\n')
# ----------------------------------------------------------------------------------------------------------------

# ----------------------------------------------------------------------------------------------------------------
def pick_player ():
    '''
    Gives an option to choose from two players X or Y
    Input: NA
    Output: Player selected to begin the game
    '''
    
    right_player = False
    
    while right_player == False:
            player = input ('\n Pick up a player from X or O : ')
            player = player.upper()

            if player == 'X' or player == 'Y':
                right_player = True
                print('\n Perfect !!! Let the game begin .. ')
            else:
                print('\n Please chose out of X or Y ')
    
    print (f'\n pick_player () : Running Parameters: player: {player} \t winner: {winner} \t tac_count: {tac_count} \n ')        
    
    return player

# -------------------------------------------------------------------------------------------------
# -------------------------------------------------------------------------------------------------


def change_turn (player):          
    '''
    Function: changes the turn of the player
    Input: Present player
    Output: The second player
    '''
    
    if player == 'X': 
        player = 'Y'
    else:
        player = 'X'
    print(f'\n change_turn: Turn of Player: {player}')
    return player
# ----------------------------------------------------------------------------------------------------------------

# ----------------------------------------------------------------------------------------------------------------


def check_win (player):
    '''
    Checks if any player has reached a winning combination
    
    Input: Player Symbol
    
    Return: True if a player has won and False otherwise
    '''
    
    if list_position.count (player) < 3: 
        return False
    
    if list_position[0] == player and list_position[1] == player and  list_position[2] == player:
        return True
    elif list_position[3] == player and list_position[4] == player and  list_position[5] == player:
        return True
    elif list_position[6] == player and list_position[7] == player and  list_position[8] == player:
        return True
    elif list_position[6] == player and list_position[4] == player and  list_position[2] == player:
        return True
    elif list_position[8] == player and list_position[4] == player and  list_position[0] == player:
        return True
    elif list_position[8] == player and list_position[5] == player and  list_position[2] == player:
        return True
    elif list_position[7] == player and list_position[4] == player and  list_position[1] == player:
        return True
    elif list_position[6] == player and list_position[3] == player and  list_position[0] == player:
        return True
    else:
        return False
    

# -------------------------------------------------------------------------------------------------
# -------------------------------------------------------------------------------------------------





retry = 'Y'


while retry:
    print_board()
    player = pick_player()

    while winner =='' and tac_count <9:
        print (f'\n run_tic_tac_toe() 1- player: {player} \t winner: {winner} \t tac_count: {tac_count} \t list_position: {list_position}\n ')        

        position_status = False
        while position_status == False:

            position = int ( input ('\n Pick a position: '))

            #print (f'position_status = {list_position}')

            if list_position[position-1] =='':
                position_status = True
                break    
            else:
                print ('\n this position is already taken. Please chose another position')


        print_board(position)

        tac_count+=1

        if check_win (player):
            winner = player
            print(f'\n Congratulations !!! player : {player} wins  the game')
            break # while winner ='' and tac_count <9:

        player = change_turn(player)

        if tac_count == 9:
            print ('Good Game but noone wins would you like to retry.  ')
            break


    retry = input ('Press Y for yes and N for no :')

    if retry.upper() == 'N':
        break


        #print (f'\n end of while loop Running Parameters: player: {player} \t winner: {winner} \t tac_count: {tac_count} \n ')        

        # Check if any player has won 
# -------------------------------------------------------------------------------------------------
# -------------------------------------------------------------------------------------------------

```
